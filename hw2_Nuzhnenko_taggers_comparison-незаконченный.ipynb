Функция принимает словоформы, написанные с нижнем регистре, и теги pymorphy, написанные в верхнем регистре. Ищет любые триграммы, составленные из словоформ и тегов.

```python
def chunker(text, element1, element2, element3):
    el1 = element1.islower()
    el2 = element2.islower()
    el3 = element3.islower()
    li_res = []
    toks = word_tokenize(text)
    c = 0
    for tok in toks:
        c += 1
        if el1:
            if tok.lower() == element1:
                if el2:
                    if toks[c] == element2:
                        if el3:
                            if toks[c + 1] == element3:
                                li_res.append(" ".join([tok, toks[c], toks[c + 1]]))
                        else:
                            if morph.parse(toks[c + 1])[0].tag.POS == element3:
                                li_res.append(" ".join([tok, toks[c], toks[c + 1]]))
                else:
                    if morph.parse(toks[c])[0].tag.POS == element2:
                        if el3:
                            if toks[c + 1] == element3:
                                li_res.append(" ".join([tok, toks[c], toks[c + 1]]))
                        else:
                            if morph.parse(toks[c + 1])[0].tag.POS == element3:
                                li_res.append(" ".join([tok, toks[c], toks[c + 1]]))
        else:
            if morph.parse(tok)[0].tag.POS == element1:
                if el2:
                    if toks[c] == element2:
                        if el3:
                            if toks[c + 1] == element3:
                                li_res.append(" ".join([tok, toks[c], toks[c + 1]]))
                        else:
                            if morph.parse(toks[c + 1])[0].tag.POS == element3:
                                li_res.append(" ".join([tok, toks[c], toks[c + 1]]))
                else:
                    if morph.parse(toks[c])[0].tag.POS == element2:
                        if el3:
                            if toks[c + 1] == element3:
                                li_res.append(" ".join([tok, toks[c], toks[c + 1]]))
                        else:
                            if morph.parse(toks[c + 1])[0].tag.POS == element3:
                                li_res.append(" ".join([tok, toks[c], toks[c + 1]]))
    return li_res
```
